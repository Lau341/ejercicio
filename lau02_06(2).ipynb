```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Leer el archivo CSV

```python
df = pd.read_csv("users.csv")
```

# Filtrar usuarios entre 45 y 60 años

```python
filtrar_edad = (df['edad'] >= 45) & (df['edad'] <= 60)
usuarios_filtrados = df[filtrar_edad]
```


# Contar por género

```python
conteo_genero = usuarios_filtrados['genero'].value_counts()
```

# Mostrar resultados

```python
print("Cantidad de usuarios entre 45 y 60 años por género:")
print(conteo_genero)
```


# Graficar

```python
sns.scatterplot(data=usuarios_filtrados, x="genero", y="edad")
plt.title("Conteo de personas que tengan entre 45 y 60 años")
plt.xlabel("Genero")
plt.ylabel("Edades")
```
